```python
#Carga los datos del archivo baby.csv


import numpy as np
import pandas as pd

data_sal  = pd.read_csv("/content/sample_data/baby.csv")
data_sal.info()
data_sal
```

1. Realiza una exploración de los datos, considerando sus tipos y los descriptivos
correspondientes

```python
# analisis exploratorio  o descriptivo
# tomar esas columnas  y calcular  la media , mediana  , desviacion standard, minima, maxima
# decir si es una variable numerica o categorica de que tipo
# complementar con grafico de distribucion  (para variables numericas ) y
#     tablas de frecuenacia para variables categoricas (si es fumador o no solo esta variable aplica)
# identificar cantidad de nulos
# identificar cantidad de duplicados

#######   CAMPOS RELEVANTES EN ANALISIS DESCIRIPTIVO ########
#Birth.Weight : Peso del bebe
#Gestational.Days: Días de gestación
#Maternal.Age: Edad de la madre
#Maternal.Height: Altura de la madre
#Maternal.Pregnancy.Weight: Peso de la madre
#Maternal.Smoker: ¿Es la madre fumadora? (True: fumadora, False: no es fumadora)

promedio_BW = data_sal["Birth.Weight"].mean()
desv_std_BW = data_sal["Birth.Weight"].std()
min_BW = data_sal["Birth.Weight"].min()
max_BW = data_sal["Birth.Weight"].max()

promedio_GD = data_sal["Gestational.Days"].mean()
desv_std_GD = data_sal["Gestational.Days"].std()
min_GD = data_sal["Gestational.Days"].min()
max_GD = data_sal["Gestational.Days"].max()

promedio_AGE = data_sal["Maternal.Age"].mean()
desv_std_AGE = data_sal["Maternal.Age"].std()
min_AGE = data_sal["Maternal.Age"].min()
max_AGE = data_sal["Maternal.Age"].max()

promedio_PESO = data_sal["Maternal.Pregnancy.Weight"].mean()
desv_std_PESO = data_sal["Maternal.Pregnancy.Weight"].std()
min_PESO = data_sal["Maternal.Pregnancy.Weight"].min()
max_PESO = data_sal["Maternal.Pregnancy.Weight"].max()

print("El promedio(La Media) general de los Peso del bebe  es  " , promedio_BW)
print("El desviacion estandard general de los Peso del bebe es  " , desv_std_BW)
print("El valor minimo de los Peso del bebe  es  " , min_BW)
print("El valor maximo de  los Peso del bebe es  " , max_BW)

print("El promedio(La Media) general de los dias de gestacion del bebe  es  " , promedio_GD)
print("El desviacion estandard general de los dias de gestacion del bebe es  " , desv_std_GD)
print("El valor minimo de dias de gestacion del bebe  es  " , min_GD)
print("El valor maximo de  dias de gestacion del bebe es  " , max_GD)

print("El promedio(La Media) general de la edad de la madre  es  " , promedio_AGE)
print("El desviacion estandard general la edad de la madre es  " , desv_std_AGE)
print("El valor minimo de  la edad de la madre es  " , min_AGE)
print("El valor maximo de  la edad de la madre es  " , max_AGE)

print("El promedio(La Media) general del peso de la madre  es  " , promedio_PESO)
print("El desviacion estandar del peso de la madre es  " , desv_std_PESO)
print("El valor minimo del peso de la madre es  " , min_PESO)
print("El valor maximo del peso de la madre madre es  " , max_PESO)


#librerias para graficar y para formatos de porcentaje
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

frec = pd.value_counts(data_sal["Maternal.Smoker"])
print("La frecuencia absoluta  de mamas fumadoras y NO fumadoras es : " , frec)

frec_df = pd.DataFrame(frec)
frec_df.columns = ["frec_smoker"]
frec_df

frec_df["Frec_relativa_%"] = 100 * frec_df["frec_smoker"] / len(data_sal)
print("La frecuencia Relativa  de mamas fumadoras y NO fumadoras es : " , frec_df)

print("La cantidad de valores Nulos en DF es : " , data_sal.isnull().any().sum())
print("Se cantidad de filas duplicadas en DF es : " , data_sal.duplicated().sum())
```

2. Considera los pesos de los bebés. ¿Se puede decir que siguen una distribución
normal? Justifica tu respuesta creando los gráficos y modelos necesarios.



```python
# crear graficos y un modelo  que nos diga como se comportan los datos.
# para crear este modelo necesitamos la media y la desviacion standard
# la media peso bebe  promedio_BW
# std peso bebe  desv_std_BW  Gestational.Days
dat_peso_baby = data_sal["Birth.Weight"]
# Graficando histograma
mu, sigma = promedio_BW, desv_std_BW # media y desvio estandar
# histograma de distribución normal.
peso = plt.hist(dat_peso_baby)
plt.ylabel('Frequencia')
plt.xlabel('Valores')
plt.title('Distribucion Peso Bebes')
plt.show()
#Conclusion , se aprecia segun grafica que es distribucion normal con una media 120 aprox
# y una desviacion standard de 18


from scipy.stats import norm, percentileofscore

x = np.linspace(min_BW , max_BW ,100)
#pdf = norm.pdf(x, loc = mu, scale = sigma )
pdf = norm.pdf(x, loc = promedio_BW, scale = desv_std_BW )

plt.hist(dat_peso_baby, bins=100, density=True, alpha = 0.5 , label = "Datos" )
plt.plot(x, pdf, color = "red" , label = "Modelo Normal")
plt.xlabel("Peso")
plt.ylabel("Densidad de Probabilidad normalizada")
plt.title("Datos vs Modelo")
plt.legend()
plt.show

```

3. Considerando lo anterior, verifica el modelo de distribución normal de los pesos para
5 valores diferentes. Compara con los datos reales.


```python
# se debe primero generar el modelo anterior y en ese modelo se puede comparar con datos reales
# y datos generados por el modelo

#Modelo generado random de 5 valores diferentes

mu, sigma = promedio_BW, desv_std_BW # media y desvio estandar
pesos_prueba = data_sal["Birth.Weight"].head()

for i in pesos_prueba:
  p_real = round(percentileofscore(dat_peso_baby, i),2)
  p_modelo = round(100*norm.cdf(i, loc=mu, scale=sigma),2)
  print(f"El percentil real para {i} es {p_real}, y el teórico es {p_modelo}")
```

5. Considerando los datos dados:
● Calcula la probabilidad de que una madre escogida al azar fume.
● A partir del valor anterior, construye un modelo de distribución binomial que
modele el experimento correspondiente a escoger 5 madres al azar y observar
cuántas fuman.
● Utilizando este modelo, simula 1000 muestras de tamaño 8, y calcula sus medias
muestrales.
 Utilizando los datos directamente, extrae 1000 muestras de tamaño 8 y calcula
sus respectivas medias muestrales.

```python
# calcular probabilidad conteo  es al azar y fume
# luego hacer un modelo binomial , la variable aleatoria es al escojer una madre al azar  saber si fuma o no

data_smoke = data_sal[data_sal['Maternal.Smoker']==True]
conteo = len(data_smoke)
probabilidad =  conteo/len(data_sal)
print("La probabilidad de que la madre fume es , " , probabilidad)


# Graficando Binomial
import scipy.stats as stats
N, p = 30, probabilidad # parametros de forma
binomial = stats.binom(N, p) # Distribución
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

```
