```python

#cargamos la informaciion a DF Pandas
import numpy as np
import pandas as pd

data_world  = pd.read_csv("/content/sample_data/worldcup2014.csv")
data_world
```

1. Considerando el experimento de seleccionar un país participante al azar, se pide
determinar las probabilidades de los siguientes sucesos:




```python
#a. pertenecer a África
data_worl_afri = data_world[data_world['continent']=='africa']
conteo = len(data_worl_afri)
print("La probabilidad de que pertenezca a africa es , " , conteo/len(data_world))

#b. pertenecer a Asia
data_worl_asia = data_world[data_world['continent']=='asia']
conteo = len(data_worl_asia)
print("La probabilidad de que pertenezca a asia  es , " , conteo/len(data_world))

#c. pertenecer a Europa
data_worl_europa = data_world[data_world['continent']=='europa']
conteo = len(data_worl_europa)
print("La probabilidad de que pertenezca a europa es , " , conteo/len(data_world))

#d. pertenecer a Norteamérica
data_worl_northamerica = data_world[data_world['continent']=='northamerica']
conteo = len(data_worl_northamerica)
print("La probabilidad de que pertenezca a northamerica es , " , conteo/len(data_world))

#e. pertenecer a Sudamérica
data_worl_southamerica = data_world[data_world['continent']=='southamerica']
conteo = len(data_worl_southamerica)
print("La probabilidad de que pertenezca a southamerica es , " , conteo/len(data_world))

#f. clasificar a segunda ronda
conteo = data_world["clasificado"].value_counts()
print("La probabildad que clasifique a segunda ronda es :" ,conteo[1]/len(data_world))

#g. ganar al menos un partido
data_worl_gana = data_world[data_world['juegos_ganados']>=1]
conteo = len(data_worl_gana)
print("La probabilidad de ganar al menos un partido es , " , conteo/len(data_world))

#h. pertenecer a África y clasificar a segunda ronda
conteo = len(data_worl_afri[data_worl_afri["clasificado"] ==1])
print("La probabilidad de pertenecer a África y clasificar a segunda ronda , " , conteo/len(data_world))

#i. pertenecer a Europa y clasificar a segunda ronda
conteo = len(data_worl_europa[data_worl_europa["clasificado"] ==1])
print("La probabilidad de pertenecer a europa  y clasificar a segunda ronda , " , conteo/len(data_world))

#j. pertenecer a Asia y clasificar a segunda ronda
conteo = len(data_worl_asia[data_worl_asia["clasificado"] ==1])
print("La probabilidad de pertenecer a asia  y clasificar a segunda ronda , " , conteo/len(data_world))

#k. pertenecer a Sudamérica y clasificar a segunda ronda
conteo = len(data_worl_southamerica[data_worl_southamerica["clasificado"] ==1])
print("La probabilidad de pertenecer a Sudamérica  y clasificar a segunda ronda , " , conteo/len(data_world))

#l. pertenecer a Norteamérica y clasificar a segunda ronda
conteo = len(data_worl_northamerica[data_worl_northamerica["clasificado"] ==1])
print("La probabilidad de pertenecer a northamerica  y clasificar a segunda ronda , " , conteo/len(data_world))

#m. pertenecer a África o clasificar a segunda ronda
conteo =len(data_world[(data_world['continent']=='africa') | (data_world['clasificado']==1)])
print("La probabilidad de pertenecer a africa o  clasificar a segunda ronda , " , conteo/len(data_world))


#n. pertenecer a Europa o clasificar a segunda ronda
conteo =len(data_world[(data_world['continent']=='europa') | (data_world['clasificado']==1)])
print("La probabilidad de pertenecer a europa o  clasificar a segunda ronda , " , conteo/len(data_world))

#o. pertenecer a Asia o clasificar a segunda ronda
conteo =len(data_world[(data_world['continent']=='asia') | (data_world['clasificado']==1)])
print("La probabilidad de pertenecer a europa o  clasificar a segunda ronda , " , conteo/len(data_world))

#p. pertenecer a Sudamérica o clasificar a segunda ronda
conteo =len(data_world[(data_world['continent']=='southamerica') | (data_world['clasificado']==1)])
print("La probabilidad de pertenecer a sudamerica o  clasificar a segunda ronda , " , conteo/len(data_world))

#q. pertenecer a Norteamérica o clasificar a segunda ronda
conteo =len(data_world[(data_world['continent']=='northamerica') | (data_world['clasificado']==1)])
print("La probabilidad de pertenecer a northamerica o  clasificar a segunda ronda , " , conteo/len(data_world))

#r. pertenecer a África y haber ganado al menos un partido
conteo = len(data_worl_afri[data_worl_afri['juegos_ganados']>=1])
print("La probabilidad de pertenecer a África y haber ganado al menos  1 partido , " , conteo/len(data_world))

#s. pertenecer a Europa y haber ganado al menos un partido
conteo = len(data_worl_europa[data_worl_europa['juegos_ganados']>=1])
print("La probabilidad de pertenecer a europa y haber ganado al menos  1 partido , " , conteo/len(data_world))

#t. pertenecer a Asia y haber ganado al menos un partido
conteo = len(data_worl_asia[data_worl_asia['juegos_ganados']>=1])
print("La probabilidad de pertenecer a asia y haber ganado al menos  1 partido , " , conteo/len(data_world))

#u. pertenecer a Sudamérica y haber ganado al menos un partido
conteo = len(data_worl_southamerica[data_worl_southamerica['juegos_ganados']>=1])
print("La probabilidad de pertenecer a sudamerica y haber ganado al menos  1 partido , " , conteo/len(data_world))

#v. pertenecer a Norteamérica y haber ganado al menos un partido

conteo = len(data_worl_northamerica[data_worl_northamerica['juegos_ganados']>=1])
print("La probabilidad de pertenecer a norteamerica y haber ganado al menos  1 partido , " , conteo/len(data_world))

#w. haber ganado al menos un partido y clasificar a segunda ronda
conteo =len(data_world[(data_world['juegos_ganados']>=1) & (data_world['clasificado']==1)])
print("La probabilidad haber ganado 1 partido  y   clasificar a segunda ronda , " , conteo/len(data_world))
```
