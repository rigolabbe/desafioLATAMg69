Crea un DataFrame con el archivo entregado.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/sample_data/resumen_resultados_astronautas.csv")
data
```

Crea la columna "IMC" usando la fórmula entregada

```python
# IMC = m / h*h
# m = peso  h = altura
#  se genera el calculo llevando  la altura de cm a metros  y se crea en la columna IMC
data["IMC"] = round(data["peso"] /  ( (data["altura"] / 100) * (data["altura"] / 100 )  ),1)
data
```

Crea la columna "promedio_evaluaciones" con el promedio de las 4 columnas
de evaluaciones. Si el IMC está entre 18.5 y 24.9, entonces el candidato
tendrá un IMC adecuado para ser astronauta. Filtra el DataFrame para
obtener solo aquellas filas donde el IMC esté dentro de los límites
entregados

```python
promedio_eval  = (data["evaluacion_fisica"] +  data["evaluacion_psicologica"]  +  data["evaluacion_adaptativa"]  +  data["evaluacion_medica"] ) / 4

data["promedio_evaluaciones"] = promedio_eval

IMC_rango = data[(data["IMC"] >= 18.5 )&(data["IMC"]<= 24.9)]
IMC_rango
```

El promedio de las 4 evaluaciones debe ser mayor o igual a 87. Filtra el
DataFrame resultante y agrega esta condición al filtro aplicado en el paso
anterior

```python
 # a lo contenido en la variable de calculo anterior ademas se filtra por  promedio de evaluaciones acalculado anteriormente
 IMC_rango =  IMC_rango[(IMC_rango["promedio_evaluaciones"] >= 87 ) ]
 IMC_rango
```

Reinicia el índice del DataFrame resultante

```python
# reindexamos los indices del resultado filtrado
IMC_rango.reset_index(drop=True, inplace=True)
IMC_rango
```

Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del
DataFrame resultante

```python
IMC_rango["califica"] = "Si"
IMC_rango
```

Exporta el DataFrame resultante a un archivo .csv separado por comas con el
nombre "astronautas_calificados.csv".


```python
# el archivo queda por default descargado con separador de comas
data.to_csv("/content/sample_data/astronautas_calificados.csv",index=False)
```
