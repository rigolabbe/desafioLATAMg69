```python
#IMPORTAMOS LAS LIBRERIAS PARA TRABAJAR
import pandas as pd
import numpy as np
# libreria para trabajar archivos
import pickle as pk

# PREGUNTA A. Carga los datos y crea un DataFrame con cada uno de ellos.
df_incidents = pd.read_pickle("/content/sample_data/incidents.pkl")
#print(df_incidents)

df_officers = pd.read_pickle("/content/sample_data/officers.pkl")
#print(df_officers)

df_subjects = pd.read_pickle("/content/sample_data/subjects.pkl")
#print(df_subjects)



```

b. Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las
 columnas que se llaman igual usando el parámetro suffixes de pd.merge().

```python
#verificar nulos por conflictos
#df_incidents.isnull().sum()
#df_officers.isnull().sum()
#df_subjects.isnull().sum()

data_merged = pd.merge(df_incidents,df_officers, on="case_number", how="inner")

# suffixes  se utiliza porque existen dos nombres de campos  case
data_merged = pd.merge(data_merged , df_subjects,  on="case_number", suffixes=("_officers", "_subjects"))
data_merged
```

b. Utiliza la siguiente lista de valores que serán considerados como nulos:
 ["na", "NA",-1, "0", "-1", "null", "n/a", "N/A", "NULL"]
 (hint: utiliza el método replace para reemplazar los valores indicados por np.nan)

```python
#   ultima pregunta del desafio se puede realizar con esta funciones

diccionario = {
    "na":"nulos",
    "NA":"nulos",
    -1:"nulos",
    "0":"nulos",
    "-1":"nulos",
    "null":"nulos",
    "n/a":"nulos",
    "N/A":"nulos",
    "NULL":"nulos"
}

df_jobs= df_jobs.replace(diccionario)
df_jobs
```

c. Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna())

```python
# borramos los daatos nan y lo dejamos en el mismo df
df_jobs = df_jobs.dropna()
df_jobs
```

d. Apartir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo
 y Máximo. (hint: Utiliza el método apply sobre la columna.)

```python

def get_salmin(registro):
  separacion = registro.split("-")
  return separacion[0]  # este 0 es porque es el primer elemento de la separacion

def get_salmax(registro):
  separacion = registro.split("-")
  return separacion[1]  # este  [1] seria el segundo elemento de la separacion

#aca  creamnos las columnas nuevas al df y asignamos los vaores segun las funciones creadas
df_jobs["Salario Minimo"] = df_jobs["Salary Estimate"].apply(get_salmin)
df_jobs["Salario Maximo"] = df_jobs["Salary Estimate"].apply(get_salmax)
df_jobs

```

 e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint:

```python
#   ultima pregunta del desafio se puede realizar con esta funciones

diccionario = {
    "10000+ employees":"Mega Empresas",
    "5001 to 10000 employees":"Grandes Empresas",
    "1001 to 5000 employees":"Medianas Empresas",
    "201 to 500 employees":"Pequeñas Empresas",
    "51 to 200 employees":"Pequeñas Grandes Empresas",
    "501 to 1000 employees":"Microempresas",
    "Unknown": "Empresas sin Información"

}

df_jobs= df_jobs.replace(diccionario)
df_jobs
```
