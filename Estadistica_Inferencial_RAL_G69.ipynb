1. Considerando las variables earn, height y age, vamos a suponer que nuestro dataset
fuera una población completa. (Prepara los datos adecuadamente).

```python
#                          PRUEBAS DE HIPOTESIS
#Hipotesis  nula es lo que tenemos como datos H0
#a. Realiza 5 pruebas con muestras de tamaño n = 25, en cada caso, con niveles
#de significancia diferentes pero inferiores a 0,1, para confirmar o rechazar las
#siguientes hipótesis considerando las alternativas. Crea para ello una fórmula
#que reciba los parámetros adecuados y resp

import numpy as np
import scipy.stats as stats

earn_dat= df['earn']
height_dat =df['height']
age_dat = df['age']

# pruebas de N=25
def prueba_hipotesis(data, H0, alternativa, nivel_confianza):
    muestra = np.random.choice(data, size=25, replace=False)
    if alternativa == '=':
        p_valor = stats.ttest_1samp(muestra, H0)[1]
    elif alternativa == '!=':
        _, p_valor = stats.ttest_1samp(muestra, H0)
    elif alternativa == '>':
        _, p_valor = stats.ttest_1samp(muestra, H0, alternative='greater')
    elif alternativa == '<':
        _, p_valor = stats.ttest_1samp(muestra, H0, alternative='less')

    if p_valor < nivel_confianza:
        resultado = f"con nivel de confianza del {(1 - nivel_confianza) * 100}%, se rechaza la hipótesis nula"
    else:
        resultado = f"con nivel de confianza del {(1 - nivel_confianza) * 100}%, se acepta la hipótesis nula"
    return resultado

# para las pruebas de hipótesis segun definicion en desafio
pruebas_earn = [(20000, '!='), (35000, '!='), (25000, '>'), (32000, '<')]
pruebas_height = [(55, '='), (68, '!='), (63, '>'), (70, '<')]
pruebas_age = [(44, '!='), (56, '!='), (51, '>'), (62, '<')]


#  pruebas  hipótesis para earn
print("Resultados para earn:")
for H0, alternativa in pruebas_earn:
    resultado = prueba_hipotesis(earn_dat, H0, alternativa, 0.1)
    print(f"H0: x {alternativa} {H0} - {resultado}")


#  pruebas hipótesis para height
print("Resultados para height:")
for H0, alternativa in pruebas_height:
    resultado = prueba_hipotesis(height_dat, H0, alternativa, 0.1)
    print(f"H0: x {alternativa} {H0} - {resultado}")


#  pruebas hipótesis para age
print("Resultados para age:")
for H0, alternativa in pruebas_age:
    resultado = prueba_hipotesis(age_dat, H0, alternativa, 0.1)
    print(f"H0: x {alternativa} {H0} - {resultado}")
```

2. Crea una función que estime la proporción de hombres en el dataset, considerando
una muestra de 50 individuos, con niveles de significancia de 0,05 y 0,01. Escribe con
palabras tus resultados ejecutar la función e interpreta. Compara con la proporción
real.


```python
def estimar_proporcion_hombres(data, muestra_size=50, nivel_confianza=0.05):
    muestra = np.random.choice(data, size=muestra_size, replace=False)
    proporcion_muestra = np.mean(muestra)

    _, p_valor = stats.ttest_1samp(muestra, 0.5)  # Comparar con la proporción real de hombres (0.5)

    if p_valor < nivel_confianza:
        resultado = f"El p-valor ({p_valor}) es menor que el nivel de confianza del {nivel_confianza * 100}%, por lo tanto, se rechaza la hipótesis nula. La proporción estimada de hombres en la muestra es {proporcion_muestra}."
    else:
        resultado = f"El p-valor ({p_valor}) es mayor que el nivel de confianza del {nivel_confianza * 100}%, por lo tanto, no hay suficiente evidencia para rechazar la hipótesis nula. La proporción estimada de hombres en la muestra es {proporcion_muestra}."

    return resultado

# Ejemplo de uso con la columna 'male' del DataFrame
proporcion_real = df['male'].mean()  # Proporción real de hombres en el dataset
resultado_05 = estimar_proporcion_hombres(df['male'], nivel_confianza=0.05)
resultado_01 = estimar_proporcion_hombres(df['male'], nivel_confianza=0.01)

print("Proporción real de hombres segun los datos:", proporcion_real)
print("Resultado para nivel de significancia 0.05:", resultado_05)
print("Resultado para nivel de significancia 0.01:", resultado_01)
```

3. Considerando el dataset como una muestra, verifica si el género (male=1 significa
‘hombre”) influye sobre el sueldo “earn” de las personas. Explica y justifica tu
procedimiento.


```python


# Separar  salarios  hombres y mujeres
earn_hombres = df[df['male'] == 1]['earn']
earn_mujeres = df[df['male'] == 0]['earn']  # si es cero es por que es mujer


# Realizar una prueba t para dos muestras independientes
t_stat, p_valor = stats.ttest_ind(earn_hombres, earn_mujeres)

# Establecer el nivel de significancia
nivel_confianza = 0.05

# Comparar el p-valor con el nivel de significancia
if p_valor < nivel_confianza:
    resultado = f"El p-valor ({p_valor}) es menor que el nivel de significancia del {nivel_confianza * 100}%, por lo tanto, se rechaza la hipótesis nula. Existe evidencia estadística para decir que el género influye en el sueldo de las personas en este dataset."
else:
    resultado = f"El p-valor ({p_valor}) es mayor que el nivel de significancia del {nivel_confianza * 100}%, por lo tanto, no hay suficiente evidencia para rechazar la hipótesis nula. No se puede concluir que el género influye en el sueldo de las personas en este dataset."

print(resultado)

```
